```python
import pandas as pd
df = pd.read_csv('/Users/edscrib/Desktop/Praktikum Webeet/Day 2 dataset/high-school-directory.csv')
df
```

```python
# cleaned column titles
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

df.columns
df
```

```python
num_schools_brooklyn = df[df["borough"] == "Brooklyn"]['dbn'].nunique()
print(num_schools_brooklyn)
```

```python
num_schools_brooklyn_grade9 = df[
    (df['borough'] == 'Brooklyn') &
    (df['grade_span_min'] <= 9) &
    (df['grade_span_max'] >= 9)
]['dbn'].nunique()

print(num_schools_brooklyn_grade9)
```

```python
schools_per_borough = df.groupby('borough')['dbn'].nunique()

print(schools_per_borough)
```

```python
avg_students_per_borough = df.groupby('borough')['total_students'].mean()

print(avg_students_per_borough)
```

```python
df.groupby('borough')['grade_span_max'].describe()
```

```python
import matplotlib.pyplot as plt

df["borough"].value_counts().plot(kind="bar")

plt.title("Number of Schools by Borough")
plt.xlabel("Borough")
plt.ylabel("Number of Schools")

plt.show()
```

```python
top_schools = df.nlargest(10, "total_students")

top_schools.plot(x="school_name", y="total_students", kind="bar")

plt.title("Top 10 Largest Schools")
plt.xlabel("School Name")
plt.ylabel("Total Students")

plt.show()
```
