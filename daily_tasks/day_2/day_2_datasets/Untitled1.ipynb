```python
import pandas as pd
df = pd.read_csv("high-school-directory.csv")
df
```

```python
df.info
```

```python
df.columns
```

```python
# cleaned column titles
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

df.columns
```

```python
#Brooklyn Schools
brooklyn_schools = df[df["borough"] == "Brooklyn"]

brooklyn_schools
```

```python
high_grade_schools = df[df["grade_span_min"] == 9]
high_grade_schools.head()
```

```python
large_schools = df[df["total_students"] > 500]
large_schools
```

```python
import matplotlib.pyplot as plt

df["borough"].value_counts().plot(kind="bar")

plt.title("Number of Schools by Borough")
plt.xlabel("Borough")
plt.ylabel("Number of Schools")

plt.show()
```

```python
df["total_students"].plot(kind="hist", bins=20)

plt.title("Distribution of Total Students")
plt.xlabel("Total Students")
plt.ylabel("Frequency")

plt.show()
```

```python
df.groupby("borough")["total_students"].mean().plot(kind="bar")

plt.title("Average Number of Students by Borough")
plt.xlabel("Borough")
plt.ylabel("Average Students")

plt.show()
```

```python
top_schools = df.nlargest(10, "total_students")

top_schools.plot(x="school_name", y="total_students", kind="bar")

plt.title("Top 10 Largest Schools")
plt.xlabel("School Name")
plt.ylabel("Total Students")

plt.show()
```
